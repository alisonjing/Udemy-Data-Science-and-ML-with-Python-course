```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Distribution Plots
```

```python
sns.distplot(tips['total_bill'], kde= False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) # two distribution plots jointed together, in between it's a scatterplot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # two distribution plots jointed together, in between it's a scatterplot
```
