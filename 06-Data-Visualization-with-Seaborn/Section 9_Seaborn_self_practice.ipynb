```python
import seaborn as sns
sns.__version__
import scipy.stats as stats
from scipy.stats import pearsonr

import matplotlib.style
import matplotlib as mpl
mpl.rcParams['patch.force_edgecolor'] = True
mpl.style.use('classic')

from warnings import filterwarnings
filterwarnings('ignore')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Distribution Plots
```

```python
sns.distplot(tips['total_bill'], kde= False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) # two distribution plots jointed together, in between it's a scatterplot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # two distribution plots jointed together, in between it's a scatterplot
```

```python
# Regression Plot
import scipy.stats as stats

sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # two distribution plots jointed together, in between it's a scatterplot
```

```python
import scipy.stats as stats
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', stat_func =stats.pearsonr)
```
