```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #allows you to actually see the plots you create inside the Jupiter notebook
```

```python
import numpy as np
x= np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y)
plt.show()
```

```python
# subplot(): the first criteria is # of rows, then # of columns, then # of plots
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OBJECT-ORIENTED METHOD
fig= plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y, 'r-')
axes1.set_title('LARGER PLOT')
axes2.plot(y,x, 'b-')
axes2.set_title('SMALLER PLOT')
```

```python
# Move the subplot axes to the bottom corner
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.45, 0.2, 0.4, 0.3])
```

```python
## Matplotlib Part 2 3/7/2022
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
plt.tight_layout() # Fixing the boundaries and over layout issues

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
## Figure Size and DPI
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2)) #dpi: dots per inch

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.jpg', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.legend(loc=10) # '0' is the best location
```
